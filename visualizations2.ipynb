- What is the distribution of the `category_name` column?

```python
import pandas as pd
import matplotlib.pyplot as plt

#data loaded into a DataFrame called 'df'
category_counts = df['category_name'].value_counts()

# Create a bar plot of the distribution
plt.figure(figsize=(8, 5))
category_counts.plot(kind='bar')
plt.title('Distribution of Category Names')
plt.xlabel('Category Name')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

```

- Which country has the most projects?

```python
import pandas as pd   
import matplotlib.pyplot as plt

df = pd.read_csv(r'C:\Users\xende\AppData\Local\Temp\Temp1_archive.zip\most_funded_feb_2023.csv')

country_counts= df['country'].value_counts()

#bar plot of country counts

plt.figure(figsize=(10, 6))
country_counts.plot(kind='bar')
plt.title('number of project by country')
plt.xlable('country')
plt.ylable('number of projects')

plt.show()
```

- What are the top 20 Kickstarter projects?

```python
import pandas as pd
import matplotlib.pyplot as plt

#data loaded into a DataFrame called 'df'
top_20_projects = df.nlargest(20, 'pledged')

#horizontal bar plot of the top 20 projects
plt.figure(figsize=(8, 6))
plt.barh(top_20_projects['name'], top_20_projects['pledged'], color='blue')
plt.title('Top 20 Kickstarter Projects')
plt.xlabel('Amount Pledged')
plt.ylabel('Project Name')
plt.gca().invert_yaxis()
plt.show()
```

- Which parent category has the most funded projects?

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv(r'C:\Users\xende\AppData\Local\Temp\Temp1_archive.zip\most_funded_feb_2023.csv')

parent_category_counts = df['parent_category'].value_counts

# Create a bar plot of the parent category counts
plt.figure(figsize=(10, 6))
parent_category_counts.plot(kind='bar')
plt.title('Number of Funded Projects by Parent Category')
plt.xlabel('Parent Category')
plt.ylabel('Number of Projects')

plt.show()
```

- Create a user-defined function that allows a user to input a category and returns a visualization of your choice.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming the data is already loaded into a DataFrame called 'df'

def plot_category(category):
    category_data = df[df['category'] == category]
    

    plt.figure(figsize=(10, 6))
    plt.bar(category_data['project_name'], category_data['funding_goal'])
    plt.title('Funding Goals for Category: {}'.format(category))
    plt.xlabel('Project')
    plt.ylabel('Funding Goal')
    plt.xticks(rotation=45)
    plt.show()


category_input = input("Enter a category: ")
plot_category(category_input)
```
